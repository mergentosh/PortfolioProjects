```python
import pandas as pd
```

```python
data=pd.read_csv(r'./pandas/fortune.csv')
```

```python
data.head(3)
```

```python
data.isnull().sum()
```

```python
data.set_index(keys="Rank",inplace=True)
```

```python
data.head(3)
```

```python
data.dtypes
```

```python
columns= ["Company","Sector","Industry","Location"]
for col in columns:
    data[col]=data[col].astype("string")
```

```python
data.dtypes
```

```python
data.nunique()
```

```python
data_1=data[["Company","Revenue","Profits"]].groupby(["Company"]).agg("max").sort_values(by="Revenue",ascending=False)
```

```python
data_1.head(3)
```

```python
data_1["Profit Margin"]=(data_1["Profits"]/data_1["Revenue"])*100
```

```python
data_1.sort_values(by="Profit Margin",ascending=False)
```

```python
data_1[data_1["Profit Margin"] > 20]
```

```python
def comments(x):
    if x > 20 : 
        return "High Margin"
    elif x > 10 and x < 20 :
        return "Healthy Margin"
    elif x > 5 and x < 10 :
        return "Normal Margin"
    elif x < 5 :
        return "Low Margin"
```

```python
lst=[*map(comments,data_1["Profit Margin"])]
```

```python
data_1["Comments"] = lst
```

```python
data_1.head()
```

```python
data_1.sort_values(by="Profit Margin",ascending=False).head(3)
```

```python
data_1.sort_values(by="Profit Margin",ascending=False)[1:].head(20)
```

```python
new_data=pd.merge(left=data_1,right=data,how="inner",on=["Revenue","Profits"])
```

```python
new_data.head()
```

```python
del new_data["Employees"]
new_data.drop(labels="Location",axis=1,inplace=True)
```

```python
new_data.head()
```

```python
new_data.set_index("Company",inplace=True)
```

```python
new_data.head(3)
```

```python
new_data[["Sector","Industry","Profit Margin","Comments"]].sort_values(by="Profit Margin",ascending=False).head(10)[1:]
```

```python
regulated_data=new_data[["Sector","Industry","Profit Margin","Comments"]].sort_values(by="Profit Margin",ascending=False).head(50)[1:]
```

```python
regulated_data.to_csv("data_fortune.csv")
```
